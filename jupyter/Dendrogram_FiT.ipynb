```python
# Count the occurrences of _rlnClassNumber for each filament_id
class_counts = dataframe.groupby('_rlnClassNumber')['Filament_ID'].value_counts()
# Unstack the counts to create a matrix where index is filament_id and columns are the counts of _rlnClassNumber
class_matrix = class_counts.unstack(fill_value=0)
countsmatrix = np.array(class_matrix)
fil_matrix = np.transpose(countsmatrix)
fil_matrix_normed = np.transpose(countsmatrix/np.sum(countsmatrix, axis=0, keepdims=True))
```

```python
pca = PCA(n_components=2)
data_pca=pca.fit_transform(embedding_value)
umap_2D = umap.UMAP(n_neighbors=100)
umap_2D1 = umap.UMAP(n_neighbors=100)
umap_2D2 = umap.UMAP(n_neighbors=100)

data_umap = umap_2D.fit_transform(embedding_value)
HLM_umap = umap_2D1.fit_transform(HLM_value)
hist_umap = umap_2D2.fit_transform(fil_matrix_normed)
```

```python
# Create a figure with 2 subplots
fig, axes = plt.subplots(2, 3, figsize=(18, 12))

# 2D KDE plot for the first dataset
sns.kdeplot(x=data_umap[:,0], y=data_umap[:,1], ax=axes[0,0], fill=True, cmap="Blues")
axes[0,0].set_title('KDE for the FiT_vector')
axes[1,0].scatter(data_umap[:,0], data_umap[:,1],s=1)
axes[1,0].set_title('scatter plot for FiT_vector')

# 2D KDE plot for the second dataset
sns.kdeplot(x=HLM_umap[:,0], y=HLM_umap[:,1], ax=axes[0,1], fill=True, cmap="Blues")
axes[0,1].set_title('KDE for the HLM')
axes[1,1].scatter(HLM_umap[:,0], HLM_umap[:,1],s=1)
axes[1,1].set_title('scatter plot for HLM')

# 2D KDE plot for the second dataset
sns.kdeplot(x=hist_umap[:,0], y=hist_umap[:,1], ax=axes[0,2], fill=True, cmap="Blues")
axes[0,2].set_title('KDE for the Relion_hist_FT')
axes[1,2].scatter(hist_umap[:,0], hist_umap[:,1],s=1)
axes[1,2].set_title('scatter plot for Relion_hist_FT')
# Display the plot
plt.savefig(path+'/kde_plot.png',dpi=300,bbox_inches='tight',pad_inches=0)
plt.show()
```

```python
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import fcluster
from scipy.cluster.hierarchy import dendrogram, linkage, leaves_list

# Assuming a specific number of clusters k
k = 3
Z1 = linkage(data_umap, 'ward',optimal_ordering=False)  # 'ward' is one method of hierarchical clustering
Z2 = linkage(hist_umap, 'ward',optimal_ordering=False)  # 'ward' is one method of hierarchical clustering
clusters1 = fcluster(Z1, k, criterion='maxclust')
clusters2 = fcluster(Z2, k, criterion='maxclust')

# Calculate the silhouette scores
silhouette_avg1 = silhouette_score(data_umap, clusters1)
silhouette_avg2 = silhouette_score(hist_umap, clusters2)

print(f"Silhouette Score for dendrogram 1: {silhouette_avg1}")
print(f"Silhouette Score for dendrogram 2: {silhouette_avg2}")
```
