```python
print(dt.now())
path_meta = p65
path = path_meta['path']
iteration = 99
path_particle=path+ path_meta['meta']
#saved_particles_embedding, saved_embedding, saved_vector_embedding, saved_TT_embedding
emb_type = 'saved_vector_embedding'
path_embedding=path+'/{}_{}.npy'.format(emb_type,iteration)
out_name='it_{}_{}_ctf'.format(iteration,emb_type)
all_value_np=np.load(path_embedding)
load_star = True
```

```python
all_value_np = all_value_np
all_value_np_mean=all_value_np[:,1:].mean(axis=1)
all_value_cls=all_value_np[:,0]
print(all_value_np_mean.shape)
if emb_type == 'saved_particles_embedding':
    embedding_value=all_value_np_mean
else:
    embedding_value=all_value_np
```

```python
# generate particle label from the 'Extract/jobXXX'
if load_star==True:
    filament_meta=EMdata.read_data_df(path_particle)
    dataframe=filament_meta.star2dataframe(relion31=True)
    helicaldic, filament_id=filament_meta.extract_helical_select(dataframe)
    filament_index=filament_meta.filament_index(helicaldic)
    labels=list(dataframe['label'].unique())
    print(labels)
    particle_label_index=[]
    all_label=dataframe['label']
    for i in range(len(all_label)):
        if all_label[i]==labels[0]: 
            particle_label_index.append(0)
        elif all_label[i]==labels[1]:
            particle_label_index.append(1)
    
    # find out the label from the 'Extract/jobXXX'
    filament_data = dataframe.groupby(['filename', 'helicaltube'])
    filament_id = list(filament_data.groups.keys())
    filament_label=list(map(lambda x: x[0].split('/')[1],filament_id))
    unique=list(set(filament_label))
    unique.sort()
    print(unique)
    filament_label_index=[]
    for i in range(len(filament_label)):
        if filament_label[i]==unique[0]: 
            filament_label_index.append(0)
        elif filament_label[i]==unique[1]:
            filament_label_index.append(1)
        elif filament_label[i]==unique[2]:
            filament_label_index.append(2)
    # check the length of the filament
    filament_length=list(map(len,filament_index))
else:
    filament_label_index=[0]*500+[1]*500
    filament_label_index = np.array(filament_label_index)
    unique = [0,1]
    #filament_label_index = 1-filament_label_index
```

```python
# seperate the particles into different clusters
dataframe_out=dataframe.drop(columns=['pid','filename','helicaltube','phi0','label'])
metadata=list(dataframe_out.columns)
metadata=[name for name in metadata]
data_out=dataframe_out.values
optics=EMdata.read_relion(path_particle).extractoptic()
n_segments = len(data_out)
for i in range(filament_cluster_number):
    locals()['cluster'+str(i)]=[]
    locals()['clusterID'+str(i)]=[]
#count=0    
for i in range(len(filament_index)):
    labels=umap_predict[i]
    locals()['clusterID'+str(labels)].append(i)
    lst=filament_index[i]
    for j in range(len(lst)):
        particle_id=int(lst[j])
        if particle_id < n_segments:
            locals()['cluster'+str(labels)].append(data_out[particle_id])
        else: 
            print('exclude')
number=0
for i in range(filament_cluster_number):
    cluster_number_count=len(locals()['cluster'+str(i)])
    print(i,cluster_number_count)
    number=number+cluster_number_count
print(number-len(data_out))
```

```python
# save the star file
import os
file_name=os.path.basename(path_particle)
output_path=os.path.dirname(path_particle)+'/'+os.path.splitext(file_name)[0]
if os.path.isdir(output_path) is False:
    os.mkdir(output_path)

for i in range(filament_cluster_number):
    cluster_name='cluster'+str(i)
    data_cluster=locals()[cluster_name]
    output=EMdata.output_star(output_path+'/FiT_'+file_name,i,data_cluster,metadata)
    output.opticgroup(optics)
    output.writecluster()
```
